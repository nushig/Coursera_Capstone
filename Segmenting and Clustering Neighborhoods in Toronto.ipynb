# Neigborhoods in Toronto

### Import and arrange data in a dataframe

```python
# Import library
from IPython.display import display_html
import requests
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 300)
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
display_html(url, raw=True)
dfs = pd.read_html(url, header = 0)
df = dfs[0]
df.head()
```

```python
df.dtypes
```

Clean the data:

```python
# drop rows with not assigned
df = df[df.Borough != 'Not assigned']
df.columns = ['PostalCode', 'Borough', 'Neighborhood']
df
```

```python
# check where Neighbourhood is not assigned
df['Neighborhood'] == 'Not assigned' 
```

```python
# Replace with Borough
df = df.replace({"Not assigned":" Queen's Park"})
```

Aggregate rows with the same Borough and PostalCode

```python
df = df.groupby(['PostalCode' , 'Borough'], as_index=False, sort=False).agg(lambda x:', '.join(x))
df
```
